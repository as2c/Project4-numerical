4. To calculate the total energy `E(t)` as a function of time, you need to integrate it over space. One way to do so is using the function below. Calculate and show the total energy `E(t)` as a function of time. How well is the energy conserved? How does this depend on the grid spacings `dx` and `dt?

```julia
# Integrate `epsilon` over space
function integrate(epsilon, n, dx)
    ni, nt = size(epsilon)
    # The boundary points are special; they have a weight of 1/2.
    # There are `ni` points overall, and the grid spacing is `dx = 2L/(ni-1)`.
    # Integrating the constant function `epsilon=1` thus gives exactly `2L`.
    E = epsilon[1, n] / 2
    for i in 2:ni-1
        E += epsilon[i, n]
    end
    E += epsilon[ni, n] / 2
    # Scale by the grid spacing
    E *= dx
    return E
end;
```

function coords(nn, ni, dt, dx, L)
    t = [n * dt for i in 0:ni, n in 0:nn]
    x = [(i * dx - L) for i in 0:ni, n in 0:nn]
    return t, x

```julia
# Define two grid functions 't', 'x' holding the coordinates
function coords(nn, ni, dt, dx, L)
    t = [n * dt for i in 0:ni, n in 0:nn]
    x = [(i * dx - L) for i in 0:ni, n in 0:nn]
    return t, x
end
```

```julia
# setting parameters
L = 1.
A = 1.
W = 0.2

nn = 401
ni = 200
dt = (4 * L) / nn
dx = 2 * L / ni
t, x = coords(nn, ni, dt, dx, L);
```

```julia
function standing(t, x, A, W)
    ni, nn = size(t)
    u = zeros(ni, nn)
    # initial conditions for first two points in time
    for n in 1:2, i in 1:ni
        u[i, n] = A * exp(-1 / 2 * sin(2Ï€ * (x[i, n] - t[i, n]))^2 / W^2)
    end
    # Boundary conditions at left and right boundary
    for n in 1:nn
        u[1, n] = 0
        u[ni, n] = 0
    end
    return u
end;
```

```julia
u = standing(t, x, A, W);
```

```julia
fig = Figure(resolution=(400, 300))
ax = Axis(fig[1, 1])
co = contourf!(vec(x), vec(t), vec(u))
Colorbar(fig[1, 2], co, label = "wave")
fig
```

```julia
# Find the solution at one interior point
function evolve_point!(u, dt, dx, n, i)
    ni, nn = size(u)
    # Finite difference for second derivative in x direction at point [i, n - 1]
    uxx = (u[i - 1, n - 1] - 2*u[i, n - 1] + u[i + 1, n - 1]) / dx^2
    # Time derivative
    # utt = (u[i, n - 2] - 2 * u[i, n - 1] + u[i, n]) / dt^2 
    # utt = uxx
    # Solve for u[i, n]
    u[i, n] = uxx * dt^2 - u[i, n - 2] + 2 * u[i, n - 1]
    return
end;
```

```julia
# Find the solution for one point in time `n`
function evolve_step!(u, dt, dx, n)
    ni, nn = size(u)
    # Apply boundary condition at left boundary
    u[1, n] = 0
    for i in 2:ni-1
        evolve_point!(u, dt, dx, n, i)
    end
    # Apply boundary condition at right boundary
    u[ni, n] = 0
    return
end;
```

```julia
function evolve!(u, dt, dx)
    ni, nn = size(u)
    # Loop over all times
    for n in 3:nn
        evolve_step!(u, dt, dx, n)
    end
    return
end;
```

```julia
evolve!(u, dt, dx)
fig = Figure(resolution=(600, 500))
ax = Axis(fig[1, 1])
co = contourf!(vec(x), vec(t), vec(u))
Colorbar(fig[1, 2], co, label = "wave")
fig
```

2. Vary the time step size `dt`. If you choose `dt` too large, the discrete system will be unstable. What is the stability limit? Why does this happen? How could you avoid this limit?

If I change nn to a number less than 400 when ni = 200, the evolution of the wave will diverge. This happens because I have set: <br>
$dt = \frac{4L}{nn}$<br>
<br>
$dx = \frac{2L}{ni}$
<br>
which means that, if $nn < 2 \times ni$ we would have:<br>
$CFL = \frac{\Delta t}{\Delta x} > 1$ and the answer would diverge. But, if $nn > 2\times ni$, $CFL < 1$ the solution will be convergent.

To avoid this limit, if we want to decrease dt further, we have to doubly decrease dx. As long as we do that, the answers won't be divergent.

3. The energy density of a scalar wave is `epsilon(t,x) = 1/2 ((d/dt u(t,x))^2 + ((d/dx u(t,x))^2)`. Calculate the energy density as a function of space and time for your solution. Plot the result.

```julia
function calc_epsilon(u, dx, dt)
    ni, nn = size(u)
    Ïµ = zeros(ni, nn)
    # inner points
    for i in 2:ni - 1, n in 2:nn - 1
       Ïµ[i, n] = 1 / 2 * ( ( (u[i + 1, n] - u[i - 1, n]) / (2 * dx))^2 + ((u[i, n + 1] - u[i, n - 1]) / (2 * dt))^2 )
    end
    
    # t = 1, x != 1,ni
    for i in 2:ni - 1
        Ïµ[i, 1] = 1 / 2 * ( ((u[i + 1, 1] - u[i - 1, 1]) / (2 * dx))^2 + ( (u[i, 2] - u[i, 1]) / dt)^2)
    end
    # Boundaries:
    # for x = 1,ni and t != 1,nn
    for n in 2:nn - 1
        Ïµ[1, n] = 1 / 2 * ( ((u[2, n] - u[1, n]) / dx)^2 + ((u[1, n + 1] - u[1, n - 1]) / (2 * dt))^2)
        Ïµ[ni, n] = 1 / 2 * ( ((u[ni, n] - u[ni - 1, n]) / dx)^2 + ((u[ni, n + 1] - u[ni, n - 1]) / (2 * dt))^2) 
    end
    # for t = 1, nn and x != 1,ni
    for i in 2:ni - 1
        Ïµ[i, 1] = 1 / 2 * ( ((u[i, 2] - u[i, 1]) / dt)^2 + ((u[i + 1, 1] - u[i - 1, 1]) / (2 * dx))^2)
        Ïµ[i, nn] = 1 / 2 * ( ((u[i, nn] - u[i, nn - 1]) / dt)^2 + ((u[i + 1, nn] - u[i - 1, nn]) / (2 * dx))^2) 
    end
    
    # for x = 1, t = 1
    Ïµ[1, 1] = 1 / 2 * ( ((u[1, 2] - u[1, 1]) / dt)^2 + ((u[2, 1] - u[1, 1]) / dx)^2)
    #for x = 1, t = nn
    Ïµ[1, nn] = 1 / 2 * ( ((u[1, nn] - u[1, nn - 1]) / dt)^2 + ((u[2, nn] - u[1, nn]) / dx)^2)
    # for x = ni, t = 1
    Ïµ[ni, 1] = 1 / 2 * ( ((u[ni, 2] - u[ni, 1]) / dt)^2 + ((u[ni, 1] - u[ni - 1, 1]) / dx)^2)
    # for x = ni, t = nn
    Ïµ[ni, nn] = 1 / 2 * ( ((u[ni, nn] - u[ni, nn - 1]) / dt)^2 + ((u[ni, nn] - u[ni - 1, nn]) / dx)^2)
    
    return Ïµ
end
```

```julia
Ïµ = calc_epsilon(u, dx, dt)
```

```julia
fig = Figure(resolution=(600, 500))
ax = Axis(fig[1, 1])
co = contourf!(vec(x), vec(t), vec(Ïµ))
Colorbar(fig[1, 2], co, label = "wave")
fig
```

4. To calculate the total energy `E(t)` as a function of time, you need to integrate it over space. One way to do so is using the function below. Calculate and show the total energy `E(t)` as a function of time. How well is the energy conserved? How does this depend on the grid spacings `dx` and `dt?

```julia
function calc_energy(Ïµ, dx)
    nt = size(Ïµ, 2)
    E = zeros(nt)
    for n in 1:nt
      E[n] = integrate(Ïµ, n, dx)
    end
    return E
end;
```

```julia
E = calc_energy(Ïµ, dx)
```

```julia
fig = Figure(resolution=(600, 500))
ax = Axis(fig[1, 1])
lines!(vec(t[1, 1:end]), vec(E), label="Energy vs Time")
axislegend(ax)
fig
```

As you can see, energy is pretty much conserved, excpet at the boundaries, But we could ignore the boundaries, since they are not exactly physically significant (we imposed the boundaries ourselves)
